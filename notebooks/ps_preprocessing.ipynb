```python
import sys
sys.path.append('../src')

import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)

from eda import get_unique_val, get_missing_values
```

```python
df_ps = pd.read_csv("../data/dataset.csv")
df_ps.head()
```

```python
df_ps.drop(['encounter_id', 'patient_id', 'hospital_id', 'icu_id', 'hospital_admit_source', 'icu_admit_source', 'icu_stay_type', 'readmission_status'], axis=1, inplace=True)
```

```python
# find missing values
df_miss = get_missing_values(df_ps)
df_miss
```

```python
# 2 ways we can drop columns that have null values above a certain threshold
# First approach is by missing count row and second is using missing percent count

# First approach
# cols_miss_val = df_ps.columns[df_ps.isnull().sum(axis=0) > 25000] # First approach

# Second approach - we drop columns that have more 50% of missing values
cols_miss_val = df_miss['col'][df_miss['pct_miss_val'] > 50.0] # Second approach
df_ps.drop(cols_miss_val, axis=1, inplace=True)
```

```python
# find number of rows where bmi, height or weight has a null
np.sum(df_ps[['bmi', 'height', 'weight']].isna().sum(axis=1) > 0)
```

```python
df_ps = df_ps[df_ps[['bmi', 'height', 'weight']].isna().sum(axis=1) == 0]
print(df_ps.shape)
```

#### Preprocessing

```python
df_ps.nunique()[df_ps.nunique() == 2].index.tolist()
```

```python
num_cat = [
    'elective_surgery',
    'apache_post_operative',
    'arf_apache',
    'gcs_unable_apache',
    'intubated_apache',
    'ventilated_apache',
    'aids',
    'cirrhosis',
    'diabetes_mellitus',
    'hepatic_failure',
    'immunosuppression',
    'leukemia',
    'lymphoma',
    'solid_tumor_with_metastasis']
```

```python
cat = df_ps.select_dtypes('O').columns.tolist()
cat
```

```python
cat_var = num_cat + cat + ['hospital_death']
cat_var
```

```python
num_var = df_ps.drop(columns=cat_var, axis=1).columns.tolist()
num_var
```

```python
for col in num_cat:
    df_ps[col] = df_ps[col].astype('Int64')

for col in num_cat:
    df_ps[col] = df_ps[col].fillna(df_ps[col].mode()[0])
```

```python
df_ps[num_var].isna().sum(axis=0).sort_values()
```

```python
num_var_gt_ts = df_ps[num_var].isna().sum(axis=0)[df_ps[num_var].isna().sum(axis=0) > 11000].index.tolist()

for col in num_var_gt_ts:
    df_ps[col] = df_ps[col].fillna(df_ps[col].mean())
```

```python
df_ps = df_ps.dropna(axis=0)
```

```python
df_ps.shape
```

```python
df_ps[cat].nunique()
```

```python
df_processed = pd.get_dummies(df_ps,
                                  prefix_sep='_',
                                  columns=cat,
                                  drop_first=False)

df_processed.reset_index(drop=True, inplace=True)
```

```python
df_processed.shape
```

```python
df_processed.columns = [col.lower() for col in df_processed.columns.tolist()]
df_processed = df_processed.loc[:, ~df_processed.columns.duplicated()]
```

```python
df_processed.head()
```

```python
int64dtype = df_processed['arf_apache'].dtype
for col in df_processed.columns.tolist():
    if df_processed[col].values.dtype == 'uint8' or df_processed[col].values.dtype == int64dtype:
        df_processed[col] = df_processed[col].astype(int)
```

```python
df_processed.dtypes
```

```python
print([col for col in df_processed.columns.tolist() if '/' in col])
df_processed.columns = [col.replace('/', '-') for col in df_processed.columns.tolist()]
```

```python
df_processed.to_csv('../data/processed_pat_sur.csv', index=False)
```
